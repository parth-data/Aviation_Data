```python
x = df['Week']
y1 = df['2019']
y2 = df['2020']
y3 = df['2021']
y4 = df['2022']
```

```python
fig, axes = plt.subplots(2, 2)

axes[0,0].plot(x, y1, '-')
axes[0,0].set_xlabel('Week')
axes[0,0].set_ylabel('2019')

axes[0,1].plot(x, y2, 'm-')
axes[0,1].set_xlabel('Week')
axes[0,1].set_ylabel('2020')

axes[1,0].plot(x, y3, 'b-')
axes[1,0].set_xlabel('Week')
axes[1,0].set_ylabel('2021')

axes[1,1].plot(x, y4, 'r-')
axes[1,1].set_xlabel('Week')
axes[1,1].set_ylabel('2022')

plt.tight_layout()
```
